```python
#Check for Stationarity
#Mean and Variance are similar for the first and second half of the data
X=ts1.OutputACCurrent
split= int(len(X)/4)
#print(split)
X1,X2,X3,X4 = X[0:split],X[split:2*split],X[2*split:3*split],X[3*split:]
#X1,X2=X[0:3000],X[3000:7074]
mean1,mean2,mean3,mean4=X1.mean(),X2.mean(),X3.mean(),X4.mean()
var1,var2,var3,var4 = X1.var(),X2.var(),X3.var(),X4.var()
print('mean1 is : %f , mean2 is : %f , mean3 is : %f , mean4 is : %f' %(mean1,mean2,mean3,mean4))
print('Variance1 is : %f , Variance2 is : %f , Variance3 is : %f , Variance4 is : %f' %(var1,var2,var3,var4))
```

```python
#ADF Test for stationarity check
from statsmodels.tsa.stattools import adfuller
X = ts1.OutputACCurrent
result = adfuller(X)
print('ADF Statistics : %f' %(result[0]))
print('p-value : %f' %(result[1]))
print('Critical Values : ')
for key,value in result[4].items():
    print('\t%s,%f' %(key,value))
```

```python
ts1['MA_3']=ts1.OutputACCurrent.rolling(3).mean()
ts1['MA_5']=ts1.OutputACCurrent.rolling(5).mean()
ts1['STD_3']=ts1.OutputACCurrent.rolling(3).std()
ts1['STD_5']=ts1.OutputACCurrent.rolling(5).std()
```

```python
#Plotting rolling statistics
plt.figure(figsize=(10,6))
plt.grid(True)
plt.plot(ts1['OutputACCurrent'],label='ACCurrent')
plt.plot(ts1['MA_3'],label='MA for 3 days')
plt.plot(ts1['MA_5'],label='MA for 5 days')
plt.plot(ts1['STD_3'],label='MSTD for 3 days')
plt.plot(ts1['STD_5'],label='MSTD for 5 dyas')
plt.legend(loc=4)
```

```python
train_data = ts1[1:len(ts)-3537]
test_data = ts1[len(ts)-3537:]
```

```python
y_hat_avg = test_data.copy()
y_hat_avg['moving_avg_forecast'] = train_data['OutputACCurrent'].rolling(60).mean().iloc[-1]
plt.figure(figsize=(16,8))
plt.plot(train_data['OutputACCurrent'], label='Train')
plt.plot(test_data['OutputACCurrent'], label='Test')
plt.plot(y_hat_avg['moving_avg_forecast'], label='Moving Average Forecast')
plt.legend(loc='best')
plt.show()
```

```python
from sklearn.metrics import mean_squared_error
from math import sqrt
rms = sqrt(mean_squared_error(test_data.OutputACCurrent, y_hat_avg.moving_avg_forecast))
ms= mean_squared_error(test_data.OutputACCurrent, y_hat_avg.moving_avg_forecast)
print('rmse is : %f' %rms)
print('mse is : %f' %ms)
```
